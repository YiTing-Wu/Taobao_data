```python
import pandas as pd
import numpy as np
```

```python
#讀取檔案
taobao_data_ori = pd.read_csv(r'C:/Users/Ting/Desktop/UserBehavior.csv',encoding='utf-8')
```

```python
#增加標題欄位
taobao_data_ori.columns = ['user_id','item_id','category_id','behavior','time']
taobao_data_ori
```

```python
#取五百萬筆資料
taobao_data_ori_part = taobao_data_ori.loc[:4999999]
taobao_data_ori_part
```

```python
#轉換時間格式
import datetime
def num_to_time(x):
    return datetime.datetime.fromtimestamp(x).strftime('%Y/%m/%d %H:%M:%S')
hr_data = taobao_data_ori_part.copy() 
hr_data['time'] = taobao_data_ori_part['time'].apply(num_to_time)
```

```python
hr_data
```

```python
#合併到dataframe刪除原始時間欄位
hr_data = pd.concat([hr_data,hr_data['time'].str.split(r' ',expand=True)],axis=1)
```

```python
hr_data = hr_data.drop(columns=['time'])
```

```python
#重新命名欄位
hr_data.rename(columns={0:'date',1:'time'}, inplace=True)
hr_data
```

```python
#以日期為準查看資料結構
hr_data.groupby('date').count()
```

```python
#選取資料較完整的日期
choose_data = hr_data[(hr_data['date'] >= '2017/11/25') & (hr_data['date'] <= '2017/12/03')]
choose_data
```

```python
#確認選取的日期
choose_data.date.unique()
```
